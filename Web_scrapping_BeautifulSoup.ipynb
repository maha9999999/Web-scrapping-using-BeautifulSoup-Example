### Introduction

In this Notebook we go over the process of web scrapping Using BeautifulSoup, through the example of  a web page of Rotten Tomatoes web site .This website provides rating of the most 100 films of all all times . We will extract  :
- The title of the film
- the score given by the audience 

### Importation

```python
from bs4 import BeautifulSoup 
```

We start making our soup by passing the file path and the parser to the BeautifulSoup constructor.We use the lxml parser in our case

```python
with open('./la_la_land.html') as file:
    soup  = BeautifulSoup(file,'lxml')
```

Then we can use the soup to extact whatever info 

### Extracting the title

We should scroll throug our html file to see whwre our tilte is placed.As we can see our title is inside a *title tag*

Now we can use the find method  by passing the 'title' tag, to extract the title

```python
title = soup.find('title')
title
```

To extract the 'La La Land (2016)' we use the contents attribute. contents provide a list of  children elements inside a given tag. In our case we have just one element which is 'La La Land (2016) - Rotten Tomatoes'

```python
title = soup.find('title').contents[0]
title
```

```python
title = title[:len(' - Rotten Tomatoes')]
title
```

### Extracting the audience score

The audience score is inside a span tag wich is inside a div tag with class 'audience-score meter'

We start by identifying the div by passing the the 'div' tag and the calss 'audience-score meter' to the find method

```python
div = soup.find('div', class_='audience-score meter')
div
```

Then we will find the first span inside this tag using find method

```python
audience_score = div.find('span')
```

```python
audience_score = audience_score.contents[0]
audience_score
```
